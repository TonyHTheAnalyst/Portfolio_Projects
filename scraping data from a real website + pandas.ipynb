# Scraping data form a real website + pandas

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_Chinese_companies'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find_all('table')
```

1. 首先是找到这个URL中你想要找到的那个table

```python
table = soup.find_all('table')[0]
```

2. 然后找到那个table中的标题 标题“th”

```python
world_titles = table.find_all('th') 
```

3. 使用for_in 遍历提取标题

```python
world_table_titles = [titles.text.strip() for titles in world_titles]

print(world_table_titles)
```

4.导入pandas 使用data frame进行标题的导入

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

5. 这一步需要将网页中表格里的数据放进df中 放在每一个对应的列下面。跟上面的步骤一样，先用for in进行遍历 将tr中的td也就是数据找出来，然后在使用一边for in 将每一行数据获取出来 存在individual_row_data当中。 然后获取使用len 获取df标题的长度，并使用df.loc[]定位length的行，在 DataFrame 的末尾添加一行新的数据，其中的值由 individual_row_data 提供。最后运行成功，输入df进行检查

```python
columns_data = table.find_all('tr') 
```

```python
for row in columns_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length]= individual_row_data
```

```python
df
```

6. 最后一步，将df的表格导出成为csv文件，给出存放路径，并去除index。这样一个csv文件就可以在Excel中打开，网上爬取的信息也成功导出。
